```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# make data
np.random.seed(1)
x = np.linspace(0, 8, 16)
y1 = 3 + 4*x/8 + np.random.uniform(0.0, 0.5, len(x))
y2 = 1 + 2*x/8 + np.random.uniform(0.0, 0.5, len(x))

# plot
fig, ax = plt.subplots()

ax.fill_between(x, y1, y2, alpha=.5, linewidth=0)
ax.plot(x, (y1 + y2)/2, linewidth=2)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

$ \lim_{h\rightarrow 0}  \frac{f(x+h)-f(x)}{h}$

```python
import numpy as np
import matplotlib.pyplot as plt

# 함수 정의
def f(x):
    return 3*x**3 + 6*x

# 미분 정의를 활용한 수치 미분 (극한 근사)
def derivative(x):
    return 9*x**2 + 6

# x 범위
x = np.linspace(-5, 5, 400) # -5부터 5까지 간격을 400으로 쪼개라

# 원래 함수와 미분 함수
y = f(x)
y_prime = derivative(x)

# 그래프 그리기
plt.figure(figsize=(8,6))
plt.plot(x, y, label=r"$f(x) = 3x^3 + 6x$", color="blue")
plt.plot(x, y_prime, label=r"$f'(x) = 9x^2 + 6$", color="red", linestyle="--")

# 좌표축 꾸미기
plt.axhline(0, color="black", linewidth=0.8)
plt.axvline(0, color="black", linewidth=0.8)
plt.title("Function and Its Derivative")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 원래 함수 정의
def f(x):
    return 3*x**2 + 7

# 미분 정의를 활용한 수치 미분 (극한 근사)
def derivative_limit(x, h=1e-5):
    return (f(x+h) - f(x)) / h

# x 구간 설정
x = np.linspace(-5, 5, 400) # -5부터 5까지 간격을 400으로 쪼개라

# 함수 값
y = f(x)

# 정의를 활용한 미분 근사
y_prime_approx = derivative_limit(x)

# 실제 해석적 미분 (검증용)
y_prime_exact = 6*x

# 그래프 그리기
plt.figure(figsize=(9,6))
plt.plot(x, y, label=r"$f(x) = 3x^3 + 6x$", color="blue")
plt.plot(x, y_prime_approx, label=r"$f'(x)$ (limit definition, approx)", color="red", linestyle="--")
plt.plot(x, y_prime_exact, label=r"$f'(x) = 9x^2 + 6$ (exact)", color="green", linestyle=":")

plt.axhline(0, color="black", linewidth=0.8)
plt.axvline(0, color="black", linewidth=0.8)
plt.title("Derivative using Limit Definition")
plt.legend()
plt.grid(True)
plt.show()
```
