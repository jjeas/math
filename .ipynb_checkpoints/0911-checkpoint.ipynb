$ \begin{bmatrix}
2 & 4 \\
6 & 2
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix}
=
\begin{bmatrix}
5 \\
7
\end{bmatrix}
\]

\[
A =
\begin{bmatrix}
2 & 4 \\
6 & 2
\end{bmatrix},
\quad
X =
\begin{bmatrix}
x \\
y
\end{bmatrix},
\quad
B =
\begin{bmatrix}
5 \\
7
\end{bmatrix}
\]

\[
AX = B \quad \Rightarrow \quad X = A^{-1}B
\]

\[
A^{-1} =
\frac{1}{(2)(2) - (6)(4)}
\begin{bmatrix}
2 & -4 \\
-6 & 2
\end{bmatrix}
=
\frac{1}{-20}
\begin{bmatrix}
2 & -4 \\
-6 & 2
\end{bmatrix}
\]

\[
X =
\frac{1}{-20}
\begin{bmatrix}
2 & -4 \\
-6 & 2
\end{bmatrix}
\begin{bmatrix}
5 \\
7
\end{bmatrix}
=
\frac{1}{-20}
\begin{bmatrix}
2\cdot 5 + (-4)\cdot 7 \\
-6\cdot 5 + 2\cdot 7
\end{bmatrix}
=
\frac{1}{-20}
\begin{bmatrix}
-18 \\
-16
\end{bmatrix}
\]

\[
X =
\begin{bmatrix}
0.9 \\
0.8
\end{bmatrix}
\quad \Rightarrow \quad
x = 0.9,\; y = 0.8
 $

```python
import numpy as np

# 계수 행렬 A
A = np.array([[2, 4],
              [6, 2]])

# 상수 벡터 B
B = np.array([5, 7])

# 해 구하기 (X = A^{-1}B)
X = np.linalg.solve(A, B)

print("해:", X)
print(f"x = {X[0]}, y = {X[1]}")
```

```python
 import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

# ---- SymPy 적분 ----
x = sp.Symbol('x', real=True)
f = 3*x**2 + 7*x + 6
a, b = 2, 5

F = sp.integrate(f, x)                 # 부정적분
area_exact = sp.integrate(f, (x, a, b))  # [2,5] 구간 면적

print("부정적분 F(x) =", F)
print(f"정확한 면적 ∫_{{{a}}}^{{{b}}} f(x) dx =", area_exact)

# ---- 리만합 직사각형 ----
N = 8   # 직사각형 개수
dx = (b - a) / N
xs = np.linspace(a, b, 400)
f_np = sp.lambdify(x, f, 'numpy')

# midpoint rule
sample_x = np.linspace(a + dx/2, b - dx/2, N)
heights = f_np(sample_x)
riemann_sum = np.sum(heights * dx)

# ---- 시각화 ----
plt.figure(figsize=(8, 5))
plt.plot(xs, f_np(xs), 'orange', label='f(x) = 3x^2+7x+6')

for xi, hi in zip(sample_x, heights):
    plt.bar(xi, hi, width=dx, alpha=0.3, edgecolor='black')

plt.axvline(a, linestyle='--', alpha=0.6)
plt.axvline(b, linestyle='--', alpha=0.6)

plt.title(f"적분 구간 [{a},{b}] Midpoint Riemann Sum (N={N})\n"
          f"리만합 ≈ {riemann_sum:.6f},  정확한 면적 = {float(area_exact):.6f}")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# ---- SymPy 적분 ----
x = sp.Symbol('x', real=True)
f = 3*x**2 + 7*x + 6
a, b = 2, 5

F = sp.integrate(f, x)                 # 부정적분
area_exact = sp.integrate(f, (x, a, b))  # [2,5] 구간 면적

print("부정적분 F(x) =", F)
print(f"정확한 면적 ∫_{{{a}}}^{{{b}}} f(x) dx =", area_exact)

# ---- 반복문으로 여러 개수의 직사각형 ----
rect_counts = [2, 3, 7, 10, 100]

xs = np.linspace(a, b, 400)
f_np = sp.lambdify(x, f, 'numpy')

# subplot을 2행 3열로 준비 (최대 6개까지)
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

for idx, N in enumerate(rect_counts):
    dx = (b - a) / N
    sample_x = np.linspace(a + dx/2, b - dx/2, N)  # midpoint rule
    heights = f_np(sample_x)
    riemann_sum = np.sum(heights * dx)

    ax = axes[idx]
    ax.plot(xs, f_np(xs), 'orange', label='f(x) = 3x^2+7x+6')

    # 직사각형 그리기
    for xi, hi in zip(sample_x, heights):
        ax.bar(xi, hi, width=dx, alpha=0.3, edgecolor='black')

    ax.axvline(a, linestyle='--', alpha=0.6)
    ax.axvline(b, linestyle='--', alpha=0.6)

    ax.set_title(f"N={N} 직사각형\n"
                 f"리만합 ≈ {riemann_sum:.4f}, 정확한 면적 = {float(area_exact):.4f}")
    ax.set_xlabel('x')
    ax.set_ylabel('f(x)')
    ax.legend()

plt.tight_layout()
plt.show()
```
