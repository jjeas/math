## decorator

```python
def hello():
    print('hello 함수 시작')
    print('hello')
    print('hello 함수 끝')
```

```python
def world():
    print('world 함수 시작')
    print('world')
    print('world 함수 끝')
```

```python
hello()
world()
```

```python
def trace(func):
    def wrapper():
        print(func.__name__,'함수 시작')
        func()
        print(func.__name__,'함수 끝')
    return wrapper
```

```python
def h():
    print('hello')
```

```python
def w():
    print('world')
```

```python
trace_hello = trace(h)
trace_hello()
trace_world = trace(w)
trace_world()
```

```python
@trace
def h():
    print('hello')
```

```python
@trace
def ww():
    print('world')
```

```python
ww()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-4, 4, 400)

fig, ax = plt.subplots()

# 두 식 (지금은 같은 직선)
ax.plot(x, (5 - x) / 3)
ax.plot(x, (5 - x) / 3)

# 직선 방정식 표시
ax.text(1, 1.6, r'$x_1 + 3x_2 = 5$')
ax.text(-3, 1.2, r'$2x_1 + 6x_2 = 10$')

# 지정한 점 (1.5, 3) 표시
ax.plot(1.5, (5-1.5)/3, 'o')
ax.text(1.5 + 0.1, 3 + 0.1, r'$(x_1, x_2) = (1.5, 3)$')

ax.set_xlim(-4, 4)
ax.set_ylim(-2, 6)
ax.axvline(color='k', linewidth=1)
ax.axhline(color='k', linewidth=1)

ax.set_xticks(list(range(-4, 5)))
ax.set_aspect('equal')
ax.grid(True, ls=':')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-4, 4, 400)

fig, ax = plt.subplots()

# 두 식 (지금은 같은 직선)
ax.plot(x, (5 - x) / 3)
ax.plot(x, (5 - 2*x) / 3)

# 직선 방정식 표시
ax.text(1, 1.6, r'$x_1 + 3x_2 = 5$')
ax.text(-3, 1.2, r'$2x_1 + 6x_2 = 10$')

# 지정한 점 (1.5, 3) 표시
ax.plot(1.5, (5-1.5)/3, 'o')
ax.text(1.5 + 0.1, 3 + 0.1, r'$(x_1, x_2) = (1.5, 3)$')

ax.set_xlim(-4, 4)
ax.set_ylim(-2, 6)
ax.axvline(color='k', linewidth=1)
ax.axhline(color='k', linewidth=1)

ax.set_xticks(list(range(-4, 5)))
ax.set_aspect('equal')
ax.grid(True, ls=':')

plt.show()
```
