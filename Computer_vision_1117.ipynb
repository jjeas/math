```python
import cv2
import numpy as np

img = data.copy()   # data = cv2.imread(...) 로 읽은 이미지

H, W, C = img.shape

# 3×2 블록 크기 계산
h_block = H // 3
w_block = W // 2

# 블록별 밝기 scale 값 (0.3 ~ 1.5)
brightness = [
    [0.5, 0.8],   # 1행
    [1.2, 0.4],   # 2행
    [1.5, 0.7]    # 3행
]

# 결과 이미지
out = img.copy()

for i in range(3):      # 3 rows
    for j in range(2):  # 2 cols

        y1 = i * h_block
        y2 = (i + 1) * h_block if i < 2 else H

        x1 = j * w_block
        x2 = (j + 1) * w_block if j < 1 else W

        block = out[y1:y2, x1:x2].astype(np.float32)

        # 밝기 조절
        block = block * brightness[i][j]
        block = np.clip(block, 0, 255).astype(np.uint8)

        # 적용
        out[y1:y2, x1:x2] = block

cv2.imwrite('./images/1117/brightness6_blocks.jpg', out)
```

```python
import cv2
import numpy as np

# 원본 이미지 읽기
img = cv2.imread("./images/1117/parrot.jpg")

# 밝기 배율 10개 (0.3 ~ 1.8 예시)
scales = np.linspace(0.3, 1.8, 10)   # 0.3, 0.47, ..., 1.8

for idx, s in enumerate(scales):
    # float32로 변환 후 밝기 조절
    out = img.astype(np.float32) * s
    
    # 값 범위 유지
    out = np.clip(out, 0, 255).astype(np.uint8)
    
    # 저장
    filename = f"{str1}/bright_{idx}.jpg"
    cv2.imwrite(filename, out)

print("생성 완료!")
```

```python
import cv2
import numpy as np
import math

# 이미지 읽기
img = cv2.imread("./images/1117/parrot.jpg")
H, W, C = img.shape

# 회전 각도 (degree)
theta = 45
rad = math.radians(theta)

# 2×2 회전행렬 (anticlockwise)
R = np.array([
    [ math.cos(rad), -math.sin(rad)],
    [ math.sin(rad),  math.cos(rad)]
])

# 출력 이미지 크기는 기존과 같게 함
out = np.zeros_like(img)

# 중심좌표
cx = W // 2
cy = H // 2

for y in range(H):
    for x in range(W):

        # (x, y) -> 중심 기준 좌표계로 이동
        xt = x - cx
        yt = y - cy

        # 회전행렬 적용
        xr, yr = R @ np.array([xt, yt])

        # 다시 원래 좌표계로 이동
        xs = int(round(xr + cx))
        ys = int(round(yr + cy))

        # 출력 이미지 범위 체크
        if 0 <= xs < W and 0 <= ys < H:
            out[y, x] = img[ys, xs]

# 저장
cv2.imwrite("./images/1117/rotated_manual.jpg", out)

print("회전 완료!")
```

```python
import cv2
import numpy as np
import math
import os

# 이미지 읽기
img = cv2.imread("./images/1117/parrot.jpg")
H, W, C = img.shape

# 회전 각도 (degree)
theta = 45
rad = math.radians(theta)

# 2×2 회전행렬 (anticlockwise)
R = np.array([
    [math.cos(rad), -math.sin(rad)],
    [math.sin(rad),  math.cos(rad)]
])

# 5회 반복하며 중심점만 달리 적용
for i in range(5):
    # 폴더 생성 (이미 존재해도 에러 없음)
    temp = f'./images/1117/temp{i}'
    os.makedirs(temp, exist_ok=True)

    # 중심점 변경
    cx = W // (i + 1)
    cy = H // (i + 1)

    # 출력 이미지 초기화
    out = np.zeros_like(img)

    # 회전 적용
    for y in range(H):
        for x in range(W):
            # (x, y) → 중심 기준 좌표계 이동
            xt = x - cx
            yt = y - cy

            # 회전행렬 적용
            xr, yr = R @ np.array([xt, yt])

            # 다시 원래 좌표계로 이동
            xs = int(round(xr + cx))
            ys = int(round(yr + cy))

            # 출력 이미지 범위 체크
            if 0 <= xs < W and 0 <= ys < H:
                out[y, x] = img[ys, xs]

    # 저장 (한 번만)
    cv2.imwrite(f"{temp}/rotated_manual_{i}.jpg", out)

print("회전 완료!")
```
