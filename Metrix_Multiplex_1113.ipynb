```python
import numpy as np
import matplotlib.pyplot as plt

# 행렬 A
A = np.array([[1, 2],
              [2, 3]])

# 고유값, 고유벡터
vals, vecs = np.linalg.eig(A)

v1 = vecs[:, 0]   # 고유벡터1
v2 = vecs[:, 1]   # 고유벡터2

# 그림
plt.figure(figsize=(6,6))
ax = plt.gca()

# 원점을 기준으로 고유벡터 표시
ax.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r', label='Eigenvector 1')
ax.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='b', label='Eigenvector 2')

# x, y 범위 설정
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)

# 격자 + 축 설정
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)
ax.grid(True)

# 고유값 텍스트 표시
plt.text(0.1, 1.2,  f"λ1 = {vals[0]:.3f}", color='r')
plt.text(0.1, -1.3, f"λ2 = {vals[1]:.3f}", color='b')

plt.title("Eigenvectors of A = [[1,2],[2,3]]")
plt.legend()
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
