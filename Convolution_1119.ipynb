```python
import cv2
import numpy as np

# -----------------------------
# 1. Convolution 직접 구현
# -----------------------------
def convolution(image, kernel):
    h, w = image.shape
    kh, kw = kernel.shape
    pad_h = kh // 2
    pad_w = kw // 2

    # 제로 패딩
    padded = np.pad(image, ((pad_h, pad_h), (pad_w, pad_w)), mode='constant')

    output = np.zeros_like(image, dtype=np.float32)

    # 컨볼루션 연산
    for y in range(h):
        for x in range(w):
            region = padded[y:y+kh, x:x+kw]
            output[y, x] = np.sum(region * kernel)

    return output


# -----------------------------
# 2. Sobel 필터 정의
# -----------------------------
sobel_x = np.array([
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1]
], dtype=np.float32)

sobel_y = np.array([
    [-1, -2, -1],
    [ 0,  0,  0],
    [ 1,  2,  1]
], dtype=np.float32)


# -----------------------------
# 3. 이미지 로드 및 처리
# -----------------------------
img = cv2.imread("./images/1117/parrot.jpg", cv2.IMREAD_GRAYSCALE)

gx = convolution(img, sobel_x)
gy = convolution(img, sobel_y)

# magnitude (엣지 강도)
mag = np.sqrt(gx**2 + gy**2)
mag = np.clip(mag, 0, 255).astype(np.uint8)

# -----------------------------
# 4. 저장
# -----------------------------
cv2.imwrite("./images/1119/sobel_x.jpg", np.abs(gx).astype(np.uint8))
cv2.imwrite("./images/1119/sobel_y.jpg", np.abs(gy).astype(np.uint8))
cv2.imwrite("./images/1119/sobel_magnitude.jpg", mag)

print("Sobel Completed!")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 평균과 표준편차
mu = 173
sigma = 5

x = np.linspace(150, 195, 1000)
pdf = norm.pdf(x, loc=mu, scale=sigma)
cdf = norm.cdf(x, loc=mu, scale=sigma)

plt.figure(figsize=(12, 5))

# PDF
plt.subplot(1, 2, 1)
plt.plot(x, pdf, label='PDF', color='blue')
plt.fill_between(x, pdf, alpha=0.2)
plt.title("확률밀도함수 (PDF)")
plt.xlabel("키 (cm)")
plt.ylabel("밀도 f(x)")
plt.grid(alpha=0.3)
plt.legend()

# CDF
plt.subplot(1, 2, 2)
plt.plot(x, cdf, label='CDF', color='green')
plt.title("누적분포함수 (CDF)")
plt.xlabel("키 (cm)")
plt.ylabel("누적 확률 F(x)")
plt.grid(alpha=0.3)
plt.legend()

plt.tight_layout()
plt.show()
```
